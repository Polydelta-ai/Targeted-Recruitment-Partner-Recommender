```python
#Import necessary libraries
import pandas as pd
import os
```

```python
#Load data and drop duplicates
df = pd.read_csv("../data/non_profit_data.csv")
url_df = pd.read_csv("../data/nccs.SC_new990_table1 (1).csv", encoding= 'unicode_escape')
_2005_df = pd.read_csv("../data/digdata.header2005b.csv", encoding= 'unicode_escape')
df.drop_duplicates(inplace = True, subset=['NAME'])
url_df.drop_duplicates(inplace = True, subset=['EIN'])
_2005_df.drop_duplicates(inplace=True, subset=['EIN'])
```

```python
#Show df summary
_2005_df.info()
```

```python
#Drop unnessecary columns and NaNs
non_profit_df = df.drop(['SEC_NAME', 'FRCD', 'RULEDATE', 'FIPS', 'PMSA', 'MSA_NECH', 'TAXPER', 'OUTREAS', 'ZFILER', 
'INCOME', 'ASSETS', 'LEVEL2', 'ACCPER', 'CASSETS', 'CFINSRC', 'CTAXPER', 'CTOTREV', 'RANDNUM', 'FILER', 'FNDNCD', 
'OUTNCCS', 'LEVEL4', 'NAICS', 'NTEECC', 'NTMAJ10', 'MAJGRPB', 'NTMAJ5'], axis=1)

non_profit_df = non_profit_df.dropna()

non_profit_df.info()
```

```python
#Reorder, rename, and drop NaNs
url_df = url_df[['EIN', 'E_Phone', "F_PrOfcr", "J_Web", "p1L01"]]

url_df = url_df.dropna()

url_df = url_df.rename(columns= {'E_Phone': "PHONE", "F_PrOfcr": "CONTACT", "J_Web": "URL", "p1L01": "MISSION"})

url_df.info()
```

```python
#Reorder, rename, and drop NaNs
_2005_df = _2005_df[['EIN', 'WEBSITE']]

_2005_df = _2005_df.dropna()

_2005_df = _2005_df.rename(columns= {'WEBSITE': 'URL'})

_2005_df.info()
```

```python
#Set indexes
_2005_df.set_index('EIN', inplace=True)
url_df.set_index('EIN', inplace=True)
non_profit_df.set_index('EIN', inplace=True)
```

```python
#Concat dfs
new_df = pd.concat([non_profit_df, url_df, _2005_df])

new_df.info()
```

```python
#Save df to CSV
new_df.to_csv('data/url-clean-non-profit.csv', index=True)
```
