```python
#Fucntion to create new columns with total number of students for each major
def add_major_student_quantity(df):
    field_of_study_percentage_list = [column for column in academic_df.columns.to_list() if column.endswith('%')]

    for major in field_of_study_percentage_list:
        
        quantity_column_name = f'{major}'.replace('%','#')
        df[quantity_column_name] = df.SIZE * df[major]
    
    return df

academic_df = add_major_student_quantity(academic_df)
```

```python
#Fucntion to create new columns with total number of students for each ethnicity
def add_student_ethnicity_quantity(df):
    student_ethnicity_percentage_list = [column for column in academic_df.columns.to_list() if column.startswith('%_')]

    for ethnicity in student_ethnicity_percentage_list:

        quantity_column_name = f'{ethnicity}'.replace('%',"#")
        df[quantity_column_name] = df.SIZE * df[ethnicity]

    return df

academic_df = add_student_ethnicity_quantity(academic_df)
```
