```python
#Import necessary libraries
import pandas as pd
```

```python
#Load data and drop duplicate schools
df = pd.read_csv("../data/academic_institutions.csv")

df.drop_duplicates(inplace = True, subset=['school.name'])
```

```python
#Drop unnecessary columns and check results
df.drop(columns=['2019.academics.program_percentage.personal_culinary', '2019.academics.program_percentage.construction', '2019.academics.program_percentage.mechanic_repair_technology',
'2019.academics.program_percentage.precision_production', '2019.academics.program_percentage.transportation'])

df.info()
```

```python
#Create df from first 40 columns and set index to 'school.name'
academic_features = df[df.columns.tolist()[:40]]

academic_features.index = df['school.name'].tolist()

academic_features.head(10)
```

```python
#Save df as CSV
academic_features.to_csv('../data/test.csv')
```
